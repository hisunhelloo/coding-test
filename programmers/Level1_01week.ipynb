[[PCCP 기출문제] 1번 / 붕대 감기](https://school.programmers.co.kr/learn/courses/30/lessons/250137)

```python
def solution(bandage, health, attacks):
    bonus_time,recover,bonus = bandage
    hp=health
    continuity=0
    attacks_count=0

    for i in range(attacks[-1][0]):
        # 몬스터 공격
        if (i+1) == attacks[attacks_count][0]:
            hp -= attacks[attacks_count][1]
            continuity = 0
            attacks_count += 1
            if hp <= 0:
                return -1
        
        # 회복
        else: 
            hp += recover
            continuity += 1
            if continuity == bonus_time:
                hp += bonus
                continuity = 0
            if hp > health:
                hp = health

    return hp

# solution([5, 1, 5], 30, [[2, 10], [9, 15], [10, 5], [11, 5]]) # 5
solution([3, 2, 7], 20, [[1, 15], [5, 16], [8, 6]])  # -1
# solution([3,10,1],100,[[1, 5], [3, 5]]) # 95
```

[[PCCE 기출문제] 9번 / 이웃한 칸](https://school.programmers.co.kr/learn/courses/30/lessons/250125)

```python
def solution(board, h, w):
    count = 0
    target = board[h][w]

    dh, dw = [0, 1, -1, 0], [1, 0, 0, -1]

    for i in range(4):
        nh, nw = h + dh[i], w + dw[i]
        if 0 <= nh < len(board) and 0 <= nw < len(board[0]):
            if board[nh][nw] == target:
                count += 1
    return count
```

```python
def solution(board, h, w):
    color=board[h][w]
    dh=[0,1,-1,0]
    dw=[1,0,0,-1]
    count=0

    for i in range(4):
        h_check=h+dh[i]
        w_check=w+dw[i]
        if 0 <= h_check < len(board) and 0 <= w_check < len(board):
            if board[h_check][w_check]==color:
                count+=1

    return count
```

[[PCCE 기출문제] 10번 / 데이터 분석](https://school.programmers.co.kr/learn/courses/30/lessons/250121)

```python
# 내 풀이
def solution(data, ext, val_ext, sort_by):
    sort_list = ["code", "date", "maximum", "remain"]
    ext_index = sort_list.index(ext)
    sort_index = sort_list.index(sort_by)
    result = []
    for i in range(len(data)):
        if data[i][ext_index] < val_ext:
            result.append(data[i])
    result.sort(key=lambda x: x[sort_index])
    return result
```

```python
# 다른 사람 풀이 1
def solution(data, ext, val_ext, sort_by):
    column_idx = {"code": 0, "date": 1, "maximum": 2, "remain": 3}
    return sorted(filter(lambda x:x[column_idx[ext]] < val_ext,data),key=lambda x:x[column_idx[sort_by]])
```

```python
# 다른 사람 풀이 2
def solution(data, ext, val_ext, sort_by):
    answer = []
    by = [ "code", "date", "maximum", "remain" ]

    for item in data:
        if item[by.index(ext)] < val_ext:
            answer.append(item)

    return sorted(answer, key=lambda x: x[by.index(sort_by)])
```

[달리기 경주](https://school.programmers.co.kr/learn/courses/30/lessons/178871)

```python
def solution(players, callings):
    count=dict()
    for i,player in enumerate(players):
        count[player]=i
    for i in callings:
        idx=count[i]
        count[i]-=1
        count[players[idx-1]]+=1
        players[idx-1],players[idx]=players[idx],players[idx-1]
    
    return players
```

[추억 점수](https://school.programmers.co.kr/learn/courses/30/lessons/176963)

```python
def solution(name, yearning, photo):
    score = {n: y for n, y in zip(name, yearning)}
    f_score=[0 for _ in photo]

    for f_idx,people in enumerate(photo):
        for person in people:
            if person in score:
                    f_score[f_idx]+=score[person]

    return f_score
```
